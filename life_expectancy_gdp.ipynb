```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('all_data.csv')
```

```python
display(df.head())
```

```python
countries = df['Country'].unique()
print(countries)
```

```python
a= 6 #rows
b= 2 #columns
c= 1 #plot couter

fig = plt.figure(figsize=[10,20])

for country in countries:
    plt.subplot(a,b,c)
    plt.title('{} GDP'.format(country))
    plt.ylabel('GDP')
    sns.scatterplot(data=df[df['Country']==country], x='Year', y='GDP')
    c=c+1
    plt.subplot(a,b,c)
    sns.scatterplot(data=df[df['Country']==country], x='Year', y='Life expectancy at birth (years)')
    plt.title('{} Life Expectancy'.format(country))
    plt.ylabel('Life Expectancy (years)')
    c=c+1
plt.tight_layout()    
plt.show()
plt.clf()
```
